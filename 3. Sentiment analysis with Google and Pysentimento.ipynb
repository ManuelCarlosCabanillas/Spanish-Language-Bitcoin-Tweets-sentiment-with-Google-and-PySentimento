```python
client = language_v1.LanguageServiceClient()
```

```python
def sample_analyze_sentiment(text_content):
    
    type_ = language_v1.Document.Type.PLAIN_TEXT
    language = "es"
    document = {"content": text_content, "type_": type_, "language": language}

    encoding_type = language_v1.EncodingType.UTF8

    response = client.analyze_sentiment(request = {'document': document, 'encoding_type': encoding_type})
   
    return response.document_sentiment.score, response.document_sentiment.magnitude
```

```python
# create three lists to findout sentiments and magnitudes
text_list = selected_tweets_df['Text']
score_list = []
magnitude_list = []
```

```python
for text in tqdm(text_list):
    score, magnitude = sample_analyze_sentiment(text)
    score_list.append(score)
    magnitude_list.append(magnitude)

selected_tweets_df['score'] = score_list
selected_tweets_df['magnitude'] = magnitude_list    
```

## PYSENTIMIENTO SENTIMENT AN√ÅLISIS

```python
from pysentimiento import create_analyzer
analyzer_sentiment = create_analyzer(task="sentiment", lang="es")
analyzer_emotion = create_analyzer(task="emotion", lang="es")
```

```python
a=analyzer_sentiment.predict("Espero que no")
```

```python
a.probas['NEU']
```

```python
b=analyzer_emotion.predict("Estoy asustado")
```

```python
b.probas['joy']
```

```python
def pysentimento_analyze(text_content):
    sentiment = analyzer_sentiment.predict(text_content)
    emotion   = analyzer_emotion.predict(text_content)
    
    sentiment_output  = sentiment.output
    sentiment_pos_probas = sentiment.probas['POS']
    sentiment_neg_probas = sentiment.probas['NEG']
    sentiment_neu_probas = sentiment.probas['NEU']
    
    emotion_output   = emotion.output
    surprise         = emotion.probas['surprise'] 
    fear             = emotion.probas['fear']
    disgust          = emotion.probas['disgust']
    sadness          = emotion.probas['sadness']
    anger            = emotion.probas['anger']
    joy              = emotion.probas['joy']
    
    
    return sentiment_output, sentiment_pos_probas, sentiment_neg_probas, sentiment_neu_probas, emotion_output, surprise, fear, disgust, sadness, anger, joy 
```

```python
# create three lists to findout sentiments and magnitudes
text_list = selected_tweets_df['Text']
sentiment_output_list = []
sentiment_pos_probas_list = []
sentiment_neg_probas_list = []
sentiment_neu_probas_list = []

emotion_output_list=[]
surprise_list=[]
fear_list=[]
disgust_list=[]
sadness_list=[]
anger_list=[]
joy_list=[]
```

```python
for text in tqdm(text_list):
    sentiment_output, sentiment_pos_probas, sentiment_neg_probas, sentiment_neu_probas, emotion_output, surprise, fear, disgust, sadness, anger, joy  = pysentimento_analyze(text)
    
    sentiment_output_list.append(sentiment_output)
    sentiment_pos_probas_list.append(sentiment_pos_probas)
    sentiment_neg_probas_list.append(sentiment_neg_probas)
    sentiment_neu_probas_list.append(sentiment_neu_probas)
    
    emotion_output_list.append(emotion_output)
    surprise_list.append(surprise)
    fear_list.append(fear)
    disgust_list.append(disgust)
    sadness_list.append(sadness)
    anger_list.append(anger)
    joy_list.append(joy)

selected_tweets_df['py_sentiment'] = sentiment_output_list
selected_tweets_df['py_sentiment_pos_probas'] = sentiment_pos_probas_list
selected_tweets_df['py_sentiment_neg_probas'] = sentiment_neg_probas_list
selected_tweets_df['py_sentiment_neu_probas'] = sentiment_neu_probas_list

selected_tweets_df['py_emotion_output'] = emotion_output_list
selected_tweets_df['py_surprise'] = surprise_list
selected_tweets_df['py_fear'] = fear_list
selected_tweets_df['py_disgust'] = disgust_list
selected_tweets_df['py_sadness'] = sadness_list
selected_tweets_df['py_anger'] = anger_list
selected_tweets_df['py_joy'] = joy_list

selected_tweets_df['py_sentiment_score'] = selected_tweets_df['py_sentiment_pos_probas'] - selected_tweets_df['py_sentiment_neg_probas']
```
